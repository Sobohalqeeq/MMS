# Definition of mms_interpol functions

```python
# Definition of mms_interpol functions
def mms_interpol():  
    
    tinterpol(fields[1], fields[0], newname=fields[1] + '_i')
    tinterpol(fields[2], fields[0], newname=fields[2] + '_i')
    tinterpol(fields[3], fields[0], newname=fields[3] + '_i')
        
    tinterpol(positions[0], fields[0], newname=positions[0] + '_i')
    tinterpol(positions[1], fields[0], newname=positions[1] + '_i')
    tinterpol(positions[2], fields[0], newname=positions[2] + '_i')
    tinterpol(positions[3], fields[0], newname=positions[3] + '_i')
        
    tinterpol(Speed_e[0], fields[0], newname=Speed_e[0] + '_i')
    tinterpol(Speed_e[1], fields[0], newname=Speed_e[1] + '_i')
    tinterpol(Speed_e[2], fields[0], newname=Speed_e[2] + '_i')
    tinterpol(Speed_e[3], fields[0], newname=Speed_e[3] + '_i')
        
    tinterpol(Speed_i[0], fields[0], newname=Speed_i[0] + '_i')
    tinterpol(Speed_i[1], fields[0], newname=Speed_i[1] + '_i')
    tinterpol(Speed_i[2], fields[0], newname=Speed_i[2] + '_i')
    tinterpol(Speed_i[3], fields[0], newname=Speed_i[3] + '_i')
    
    tinterpol(E[0], fields[0], newname=E[0] + '_i')
    tinterpol(E[1], fields[0], newname=E[1] + '_i')
    tinterpol(E[2], fields[0], newname=E[2] + '_i')
    tinterpol(E[3], fields[0], newname=E[3] + '_i')
    
    tinterpol(des_prestensor[0], fields[0], newname=des_prestensor[0] + '_i')
    tinterpol(des_prestensor[1], fields[0], newname=des_prestensor[1] + '_i')
    tinterpol(des_prestensor[2], fields[0], newname=des_prestensor[2] + '_i')
    tinterpol(des_prestensor[3], fields[0], newname=des_prestensor[3] + '_i')
    
    tinterpol(densitye[0], fields[0], newname=densitye[0] + '_i')
    tinterpol(densitye[1], fields[0], newname=densitye[1] + '_i')
    tinterpol(densitye[2], fields[0], newname=densitye[2] + '_i')
    tinterpol(densitye[3], fields[0], newname=densitye[3] + '_i')

    tinterpol(densityi[0], fields[0], newname=densityi[0] + '_i')
    tinterpol(densityi[1], fields[0], newname=densityi[1] + '_i')
    tinterpol(densityi[2], fields[0], newname=densityi[2] + '_i')
    tinterpol(densityi[3], fields[0], newname=densityi[3] + '_i')    
    
    return
```

```python
# import the mms_interpol function
mms_interpol_vars = mms_interpol()
```

# Extract the vectors

```python
# extract the vector
# B magnetic field (nT)
b1 = datab1[:, 0:3]
b2 = datab2[:, 0:3]
b3 = datab3[:, 0:3]
b4 = datab4[:, 0:3]
B=(b1+b2+b3+b4)/4

# P positions (Km)
p1 = p1[:, 0:3]
p2 = p2[:, 0:3]
p3 = p3[:, 0:3]
p4 = p4[:, 0:3]
p=(p1+p2+p3+p4)/4

# Vi ions (km/s)
vi1 = datavi1[:, 0:3]
vi2 = datavi2[:, 0:3]
vi3 = datavi3[:, 0:3]
vi4 = datavi4[:, 0:3]
vi=(vi1+vi2+vi3+vi4)/4

# Ve electrons (km/s)
ve1 = datave1[:, 0:3]
ve2 = datave2[:, 0:3]
ve3 = datave3[:, 0:3]
ve4 = datave4[:, 0:3]
ve=(ve1+ve2+ve3+ve4)/4

# E electron density (p/cm3)
de1 = datadensitye1
de2 = datadensitye2
de3 = datadensitye3
de4 = datadensitye4
de= (de1+de2+de3+de4)/4

# Di Ion density (p/cm3)
di1 = datadensityi1
di2 = datadensityi2
di3 = datadensityi3
di4 = datadensityi4
di= (di1+di2+di3+di4)/4

# E Electric field
E1 = dataE1[:,0:3]
E2 = dataE2[:,0:3]
E3 = dataE3[:,0:3]
E4 = dataE4[:,0:3]
E= (E1+E2+E3+E4)/4
```

# Definition of Current density

```python
# J
# We calculate J from the FPI particle moments using J=qene(Vi−Ve)
# current units in nA/m2 (de in p/cm3 converted in p/m3 by cm3tom3=1.e6 and velocity in km/s 
# converted in m/s by kmtom= 1.e3 lead to A/m2 then nA/m2 by multiplying by 1.e9)

jx1=cm3tom3*qe*de1*(datavi1[:,0]-datave1[:,0])*kmtom*(1e9)
jx2=cm3tom3*qe*de2*(datavi2[:,0]-datave2[:,0])*kmtom*(1e9)
jx3=cm3tom3*qe*de3*(datavi3[:,0]-datave3[:,0])*kmtom*(1e9)
jx4=cm3tom3*qe*de4*(datavi4[:,0]-datave4[:,0])*kmtom*(1e9)

jy1=cm3tom3*qe*de1*(datavi1[:,1]-datave1[:,1])*kmtom*(1e9)
jy2=cm3tom3*qe*de2*(datavi2[:,1]-datave2[:,1])*kmtom*(1e9)
jy3=cm3tom3*qe*de3*(datavi3[:,1]-datave3[:,1])*kmtom*(1e9)
jy4=cm3tom3*qe*de4*(datavi4[:,1]-datave4[:,1])*kmtom*(1e9)

jz1=cm3tom3*qe*de1*(datavi1[:,2]-datave1[:,2])*kmtom*(1e9)
jz2=cm3tom3*qe*de2*(datavi2[:,2]-datave2[:,2])*kmtom*(1e9)
jz3=cm3tom3*qe*de3*(datavi3[:,2]-datave3[:,2])*kmtom*(1e9)
jz4=cm3tom3*qe*de4*(datavi4[:,2]-datave4[:,2])*kmtom*(1e9)

jx=(jx1+jx2+jx3+jx4)/4
jy=(jy1+jy2+jy3+jy4)/4
jz=(jz1+jz2+jz3+jz4)/4

jt=np.array([jx,jy,jz])
jpart=np.transpose(jt)

jpart[:,0]=jt[0]
jpart[:,1]=jt[1]
jpart[:,2]=jt[2]

store_data('Jpart', data={'x':times, 'y': jpart[:,0:3]})
options('Jpart', 'ytitle', 'Jpart [nA/m2]')
options('Jpart', 'Color', ['b','g','r'])
options('Jpart','legend_names',['Jpart_x','Jpart_y','Jpart_z'])
#tplot('Jpart')
```
